```python
#download talk schedule from pycon website

import random
import pprint
import requests
import csv
import pandas as pd
import numpy as np

from requests_html import HTMLSession
session = HTMLSession()
r = session.get('https://us.pycon.org/2018/schedule/talks/list/')
```

```python
a = r.html.find('#content', first=True)
a.text
```

```python
#cleaning the data
clean_talks = a.text.split('\n')
clean_talks
```

```python
clean_talks[3:7]
```

```python
#formatting to a more readable structure
talks = []
for i in range(3, len(clean_talks), 4):
    talk = {}
    talk['Title'] = clean_talks[i]
    talk['Speaker'] = clean_talks[i+1]
    data = clean_talks[i+2].split(' in ')
    talk['Time'] = data[0]
    talk['Room'] = data[1]
    talk['Description'] = clean_talks[i+3]
    talks.append(talk)
```

```python
talks[0]
```

```python
pp = pprint.PrettyPrinter(indent = 4)
pp.pprint(random.choice(talks))
```

```python
header = ['Title', 'Time', 'Speaker', 'Room', 'Description']

with open('Pycon_2018_talks.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = header)
    writer.writeheader()
    for t in talks:
        writer.writerow(t)
```

```python
#define metadata function
def summary(talks):
    print('Number of talks', len(talks))
    average_description = np.mean([len(x['Description']) for x in talks])
    print('Average length of description', average_description)
    average_title = np.mean([len(x['Title']) for x in talks])
    print('Average length of title', average_title)
```

```python
summary(talks)
```
